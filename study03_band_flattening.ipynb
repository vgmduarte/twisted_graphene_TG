Vamos plotar em um só gráfico as bandas do TBG se aproximando do ângulo do mágico. A inspiração é o vídeo do paper do time do Plabo Jarrilo-Herrero.

```python
pp=[10,15,20,30]
for p in pp:
    print(twist_angle(p,1)*180/np.pi,4*number_of_sublattice_positions(p,1))
```

```python
tbgs={}
for p in pp:
    g=TwistedBilayerGraphene(p,1)
    g.calc_hops(3.64)
    g.set_kpath(g.path_KGMKp,[20,20,20])
    g.ell-=g.ell[g.kticks[1]]
    g.set_hamiltonian(1.0,0.0)
    g.calc_bands(12,0.0)
    tbgs[p]=g
```

```python
colors=['tab:blue','tab:orange','tab:red','tab:green']
for p,color in zip(pp,colors):
    g=tbgs[p]
    plt.plot(g.ell,1000*g.bands,'.',ms=3,color=color)

# plt.grid(True)
plt.xlabel('$\\ell$ [$Å^{-1}$]')
plt.ylabel('Energy [meV]')
plt.ylim((-100,100))
plt.show()
```
