```python
g=TwistedBilayerGraphene(1,0)
```

```python
r1=g.r[g.r[:,2]==0.0]
```

```python
a1,a2=g.L1,g.L2
```

```python
a=2.46

height=3.5
ratio=2
width=ratio*height

plt.figure(figsize=(width,height))

#pontos da rede
for m in range(-10,10):
    for n in range(-10,10):
        r=r1+m*g.L1+n*g.L2
        x=r[:,0]
        y=r[:,1]
        plt.plot(x/a,y/a,'.',color='tab:blue')

#borda da célula unitária
vertices=np.array([np.zeros(3),a1,a1+a2,a2,np.zeros(3)])-(a1+a2)/3
x=vertices[:,0]
y=vertices[:,1]
plt.plot(x/a,y/a,'k--',lw=1.5)
        
#vetores primitivos
origin=(a1+a2)/6
x0=origin[0]
y0=origin[1]
vectors=np.array([[x0,y0,a1[0],a1[1]],[x0,y0,a2[0],a2[1]]])/a #[xi,yi,xf,yf]
X,Y,U,V=zip(*vectors)
plt.quiver(X,Y,U,V,angles='xy', scale_units='xy', scale=1, color=['black', 'black'])
plt.annotate('$\\mathbf{a}_1$',(a1[0]/a+0.2,a1[1]/a-0.2))
plt.annotate('$\\mathbf{a}_2$',(a2[0]/a+0.2,a2[1]/a+0.1))
    
ymax=2
xmax=ratio*ymax
plt.xlim((-xmax,xmax))
plt.ylim((-ymax,ymax))

plt.xlabel('$x/a$')
plt.ylabel('$y/a$')

plt.grid(True,alpha=0.2)

plt.savefig('Fig01_monolayer_graphene_unit_cell.svg')

plt.show()
```
