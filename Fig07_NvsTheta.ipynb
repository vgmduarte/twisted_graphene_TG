```python
pp=np.arange(1,101)
qq=np.arange(1,101)
thetas=[]
Ns=[]
for p in pp:
    for q in qq:
        if np.gcd(p,q)==1:
            theta=twist_angle(p,q)
            N=4*number_of_sublattice_positions(p,q)
            thetas.append(theta)
            Ns.append(N)
thetas=np.array(thetas)
Ns=np.array(Ns)
```

```python
tt=np.array([twist_angle(p,1) for p in [1,5,10,20,30]])
nn=np.array([4*number_of_sublattice_positions(p,1) for p in [1,5,10,20,30]])
```
