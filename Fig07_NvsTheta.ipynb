```python
pp=np.arange(1,101)
qq=np.arange(1,101)
thetas=[]
Ns=[]
for p in pp:
    for q in qq:
        if np.gcd(p,q)==1:
            theta=twist_angle(p,q)
            N=4*number_of_sublattice_positions(p,q)
            thetas.append(theta)
            Ns.append(N)
thetas=np.array(thetas)
Ns=np.array(Ns)
```

```python
plt.figure(figsize=(4.9,3.5))

o=np.linspace(0,np.pi/3,10000)
plt.plot(o*180/np.pi,1/np.sin(o/2)**2,'--',lw=2,color='tab:orange')

plt.plot(thetas*180/np.pi,Ns,'.',color='tab:blue')

plt.xlim((0,60))
plt.ylim((0,7000))

plt.xlabel('TBG twist angle [$^\circ$]')
plt.ylabel('Number of atoms in TBG unit cell')

plt.savefig('Fig07_NvsTheta.svg')

plt.show()
```
