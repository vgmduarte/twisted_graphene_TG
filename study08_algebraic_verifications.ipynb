# L2 = rot 60 L1

```python
a=sp.Symbol('a')
p,q=sp.symbols('p, q',integer=True)
```

```python
sp.fraction
```

```python
sp.fraction()
```

```python
sp.sqrt(3)/2
```

```python
a1=a*sp.Matrix([sp.sqrt(3)/2,sp.Rational(-1,2),0])
a2=a*sp.Matrix([sp.sqrt(3)/2,sp.Rational(1,2),0])
```

## primeiro caso: q divisível por 3

```python
L1=(p+q/3)*a1+(q/3)*a2
```

```python
#60 degree rotation 
rot=sp.Matrix([[sp.Rational(1,2),-sp.sqrt(3)/2,0],[sp.sqrt(3)/2,sp.Rational(1,2),0],[0,0,1]])
sp.simplify(rot*L1)
```

```python
sp.simplify((-q/3)*a1+(p+2*q/3)*a2)
```

## segundo caso: q não divisível por 3

```python
L1=p*a1+(p+q)*a2
```

```python
#60 degree rotation 
rot=sp.Matrix([[sp.Rational(1,2),-sp.sqrt(3)/2,0],[sp.sqrt(3)/2,sp.Rational(1,2),0],[0,0,1]])
sp.simplify(rot*L1)
```

```python
sp.simplify((-p-q)*a1+(2*p+q)*a2)
```

# G1 e G2

## primeiro caso: q divisível por 3

```python
L1=(p+q/3)*a1+(q/3)*a2
L2=sp.simplify(rot*L1)
```

```python
z=sp.Matrix([0,0,1])
LL=sp.simplify(L1.cross(L2).dot(z))
LL
```

```python
z=sp.Matrix([0,0,1])
G1=sp.simplify(2*sp.pi*(L2.cross(z))/LL)
G1
```

```python
G2=sp.simplify(2*sp.pi*(z.cross(L1))/LL)
G2
```

## segundo caso: q não divisível por 3

```python
L1=p*a1+(p+q)*a2
L2=sp.simplify(rot*L1)
```

```python
z=sp.Matrix([0,0,1])
LL=sp.simplify(L1.cross(L2).dot(z))
LL
```

```python
z=sp.Matrix([0,0,1])
G1=sp.simplify(2*sp.pi*(L2.cross(z))/LL)
G1
```

```python
G2=sp.simplify(2*sp.pi*(z.cross(L1))/LL)
G2
```
