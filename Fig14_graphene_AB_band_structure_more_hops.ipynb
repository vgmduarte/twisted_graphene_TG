```python
def grapheneAB(max_interlayer_distance):
    g=TwistedBilayerGraphene(0,1)
    g.calc_hops(max_interlayer_distance)
    g.set_kpath(g.path_GMKG,[100,100,100])
    g.set_hamiltonian(1.0,0.0)
    g.calc_bands(4,0.0)
    return g
```

```python
from ipywidgets import interact
```

```python
@interact(max_interlayer_distance=(0.1,8.7,0.1))
def plot(max_interlayer_distance=3.36):
    g=grapheneAB(max_interlayer_distance)

    plt.plot(g.ell,g.bands/2.8922,color='tab:blue')

    plt.grid(True)
    plt.xlim((g.ell[0],g.ell[-1]))
    plt.xticks(g.ell[g.kticks],['$\mathbf{\Gamma}$','$\mathbf{M}$','$\mathbf{K}$','$\mathbf{\Gamma}$'])
    plt.ylim((-3.2,3.2))

    plt.show()
```

```python
gg={}
for i,d in zip([1,2,3,4],[3.64,4.16,4.393,5.421]):
    gg[i]=grapheneAB(d)
```

```python
g1=grapheneAB(3.36)
g2=grapheneAB(3.64)
g3=grapheneAB(5.421)
```

```python
plt.plot(g1.ell,g1.bands/2.8922,'--',color='tab:blue',lw=1.8)
plt.plot(g2.ell,g2.bands/2.8922,'-.',color='tab:orange',lw=1.2)
plt.plot(g3.ell,g3.bands/2.8922,color='tab:green',lw=1.0)

plt.grid(True)
plt.xlim((g1.ell[0],g1.ell[-1]))
plt.xticks(g1.ell[g1.kticks],['$\mathbf{\Gamma}$','$\mathbf{M}$','$\mathbf{K}$','$\mathbf{\Gamma}$'])
plt.ylim((-3.2,3.2))
plt.ylabel('$\\varepsilon(\mathbf{k})/t_1$')

plt.savefig('Fig14_graphene_AB_band_structure_more_hops.svg')

plt.show()
```
