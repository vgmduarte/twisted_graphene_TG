```python
def plot_vectors(inipts,endpts):
    plt.quiver(inipts[:,0],inipts[:,1],endpts[:,0],endpts[:,1],scale=1,scale_units='xy',angles='xy',units='xy')
    plt.axis('equal')
    plt.xlim((min(endpts.min()-1,-1)),max(endpts.max()+1,1))
    plt.ylim((min(endpts.min()-1,-1)),max(endpts.max()+1,1))
#     print(endpts.min()-1,endpts.max())
```

```python
g=TwistedBilayerGraphene(1,0)

#plot polygon
vertices=np.array([g.K[0],g.Kp[0],g.K[1],g.Kp[1],g.K[2],g.Kp[2],g.K[0]])
plt.plot(vertices[:,0],vertices[:,1],'k--')

#plot vectors
plot_vectors(np.array([[0,0],[0,0],[0,0],[0,0]]),np.array([g.L1,g.L2,g.G1,g.G2]))

plt.xlabel('x [Å]')
plt.ylabel('y [Å]')
plt.show()
```
