```python
def graphene_AA(max_interlayer_hopping_distance):
    g=TwistedBilayerGraphene(1,0)
    g.calc_hops(max_interlayer_hopping_distance)
    g.set_kpath(g.path_GMKG,[100,100,100])
    g.set_hamiltonian(1.0,0.0)
    g.calc_bands(4,0.0)
    return g
```

```python
from ipywidgets import interact
```

```python
m=6.161295933314082

@interact(d=(0.1,8.7,0.1))
def plot(d=3.4):
    g=graphene_AA(d)

    plt.plot(g.ell,g.bands/2.8922,color='tab:blue')
    
    plt.plot(g.ell,0.11+m*(g.ell-2.326)/2.8922,'--',lw=1.2,color='tab:orange')
    plt.plot(g.ell,-0.11+m*(g.ell-2.326)/2.8922,'--',lw=1.2,color='tab:orange')
    plt.plot(g.ell,-0.11-m*(g.ell-2.326)/2.8922,'--',lw=1.2,color='tab:orange')
    plt.plot(g.ell,0.11-m*(g.ell-2.326)/2.8922,'--',lw=1.2,color='tab:orange')
    
    plt.grid(True)
    plt.xticks(g.ell[g.kticks],['$\mathbf{\Gamma}$','$\mathbf{M}$','$\mathbf{K}$','$\mathbf{\Gamma}$'])
    plt.xlim((g.ell[0],g.ell[-1]))
    plt.ylabel('$\\varepsilon(\mathbf{k})/t_1$')
    plt.ylim((-3.2,3.2))

    plt.show()
```

```python
g1=graphene_AA(3.4)
```

```python
g2=graphene_AA(5.5)
```

```python
g=g1
plt.plot(g.ell,g.bands/2.8922,color='tab:blue',lw=1.5)

g=g2
plt.plot(g.ell,g.bands/2.8922,color='tab:orange',lw=1)

plt.grid(True)
plt.xticks(g.ell[g.kticks],['$\mathbf{\Gamma}$','$\mathbf{M}$','$\mathbf{K}$','$\mathbf{\Gamma}$'])
plt.xlim((g.ell[0],g.ell[-1]))
plt.ylabel('$\\varepsilon(\mathbf{k})/t_1$')
plt.ylim((-3.2,3.2))

plt.show()
```
