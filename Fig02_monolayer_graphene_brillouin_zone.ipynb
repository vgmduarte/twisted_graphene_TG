```python
g=TwistedBilayerGraphene(1,0)
```

```python
b1=g.G1
b2=g.G2
```

```python
a=2.46

height=3.5
ratio=1.5
width=ratio*height

plt.figure(figsize=(width,height))

#borda da zona de brillouin
vertices=np.array([g.K[0],g.Kp[0],g.K[1],g.Kp[1],g.K[2],g.Kp[2],g.K[0]])
x=vertices[:,0]
y=vertices[:,1]
plt.plot(x*a,y*a,'--',lw=1.5,color='gray')

#vetores primitivos
origin=np.zeros(3)
x0=origin[0]
y0=origin[1]
vectors=np.array([[x0,y0,b1[0],b1[1]],[x0,y0,b2[0],b2[1]]])*a #[xi,yi,xf,yf]
X,Y,U,V=zip(*vectors)
plt.quiver(X,Y,U,V,angles='xy', scale_units='xy', scale=1, color=['black', 'black'])
plt.annotate('$\\mathbf{b}_1$',(b1[0]*a-1.8,b1[1]*a+0.5))
plt.annotate('$\\mathbf{b}_2$',(b2[0]*a-2.0,b2[1]*a-1.0))

#pontos de alta simetria
plt.plot(g.Gamma[0]*a,g.Gamma[1]*a,'o',ms=4,color='tab:purple')
for i in range(len(g.K)):
    plt.plot(g.K[i][0]*a,g.K[i][1]*a,'o',ms=4,color='tab:blue')
for i in range(len(g.Kp)):
    plt.plot(g.Kp[i][0]*a,g.Kp[i][1]*a,'o',ms=4,color='tab:orange')
for i in range(len(g.M)):
    plt.plot(g.M[i][0]*a,g.M[i][1]*a,'o',ms=4,color='tab:red')
for i in range(len(g.Mp)):
    plt.plot(g.Mp[i][0]*a,g.Mp[i][1]*a,'o',ms=4,color='tab:green')
plt.annotate('$\\mathbf{K}$',(g.K[0][0]*a,g.K[0][1]*a-1.2),color='tab:blue')
plt.annotate('$\\mathbf{K}$',(g.K[1][0]*a+0.4,g.K[1][1]*a),color='tab:blue')
plt.annotate('$\\mathbf{K}$',(g.K[2][0]*a-1.2,g.K[2][1]*a),color='tab:blue')
plt.annotate("$\\mathbf{K}'$",(g.Kp[0][0]*a+0.4,g.Kp[0][1]*a-1.0),color='tab:orange')
plt.annotate("$\\mathbf{K}'$",(g.Kp[1][0]*a-0.3,g.Kp[1][1]*a+0.4),color='tab:orange')
plt.annotate("$\\mathbf{K}'$",(g.Kp[2][0]*a-1.2,g.Kp[2][1]*a-1.0),color='tab:orange')
plt.annotate("$\\mathbf{M}$",(g.M[0][0]*a+0.5,g.M[0][1]*a-0.8),color='tab:red')
plt.annotate("$\\mathbf{M}$",(g.M[1][0]*a+0.6,g.M[1][1]*a),color='tab:red')
plt.annotate("$\\mathbf{M}$",(g.M[2][0]*a-1.5,g.M[2][1]*a-0.3),color='tab:red')
plt.annotate("$\\mathbf{M}'$",(g.Mp[0][0]*a+0.3,g.Mp[0][1]*a),color='tab:green')
plt.annotate("$\\mathbf{M}'$",(g.Mp[1][0]*a-1.6,g.Mp[1][1]*a),color='tab:green')
plt.annotate("$\\mathbf{M}'$",(g.Mp[2][0]*a-1.6,g.Mp[2][1]*a-0.8),color='tab:green')
plt.annotate('$\\mathbf{\Gamma}$',(-0.8,0),color='tab:purple')


ymax=6.5
xmax=ratio*ymax
plt.xlim((-xmax,xmax))
plt.ylim((-ymax,ymax))

plt.xlabel('$k_x a$')
plt.ylabel('$k_y a$')

plt.grid(True,alpha=0.2)

plt.savefig('Fig02_monolayer_graphene_brillouin_zone.svg')

plt.show()
```
