```python
height=3.5
ratio=2
width=ratio*height

plt.figure(figsize=(width,height))

plt.subplot(121)

g=TwistedBilayerGraphene(4,1)
r1=g.r[g.r[:,2]==0.0]
r2=g.r[g.r[:,2]!=0.0]

plt.title(f'$\\theta(4,1)\\approx${g.theta*180/np.pi:.2f}$^\circ$')

for m in range(-4,5):
    for n in range(-4,5):
        v1=r1+m*g.L1+n*g.L2
        v2=r2+m*g.L1+n*g.L2
        x1,y1=v1[:,0],v1[:,1]
        x2,y2=v2[:,0],v2[:,1]
        plt.plot(x1,y1,'.',color='tab:blue',ms=1)
        plt.plot(x2,y2,'.',color='tab:blue',ms=1)
        
vertices=np.array([np.zeros(3),g.L1,g.L1+g.L2,g.L2,np.zeros(3)])
x=vertices[:,0]
y=vertices[:,1]
plt.plot(x,y,'k--',lw=1.5)

plt.xlabel('$x$ [Å]')
plt.ylabel('$y$ [Å]')

ymax=50
xmax=ratio*ymax/2
plt.xlim((-xmax,xmax))
plt.ylim((-ymax,ymax))

plt.subplot(122)


g=TwistedBilayerGraphene(7,3)
r1=g.r[g.r[:,2]==0.0]
r2=g.r[g.r[:,2]!=0.0]

plt.title(f'$\\theta(7,3)\\approx${g.theta*180/np.pi:.2f}$^\circ$')

for m in range(-4,5):
    for n in range(-4,5):
        v1=r1+m*g.L1+n*g.L2
        v2=r2+m*g.L1+n*g.L2
        x1,y1=v1[:,0],v1[:,1]
        x2,y2=v2[:,0],v2[:,1]
        plt.plot(x1,y1,'.',color='tab:orange',ms=1)
        plt.plot(x2,y2,'.',color='tab:orange',ms=1)
        
vertices=np.array([np.zeros(3),g.L1,g.L1+g.L2,g.L2,np.zeros(3)])
x=vertices[:,0]
y=vertices[:,1]
plt.plot(x,y,'k--',lw=1.5)

plt.xlabel('$x$ [Å]')

ymax=50
xmax=ratio*ymax/2
plt.xlim((-xmax,xmax))
plt.ylim((-ymax,ymax))

plt.savefig('Fig08_moire_patterns.png',dpi=300)
        
plt.show()
```
