```python
def plot_bands_with_cmap(bands,ell,weights,cmap):
    vmin=np.min(weights)
    vmax=np.max(weights)
    for i in range(bands.shape[1]):
        plt.scatter(ell,bands[:,i],c=weights[:,i],cmap=cmap,marker='.',vmin=vmin,vmax=vmax,s=0.2)
```
